```python
for features in continous_data:
    data_log=dataset.copy()
    if 0 in data_log[features].unique():
        pass
    else:
        data_log[features]=np.log(data_log[features])
        data_log['SalePrice']=np.log(data_log['SalePrice'])
        plt.scatter(data_log[features],data_log['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
for features in continous_data:
    data_outliers=dataset.copy()
    if 0 in data_outliers[features].unique():
        pass
    else:
        data_outliers[features]=np.log(data_outliers[features])
        data_outliers.boxplot(column=features)
        plt.show()
        
        
```
